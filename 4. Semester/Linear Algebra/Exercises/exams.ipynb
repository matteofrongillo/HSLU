```python
print(f"ln(2): {np.log(2)}")

partial_sum = 0
for k in range(1,21):
    partial_sum += ((-1)**(k-1))/k
print("Series:", partial_sum)

percentace_deviation = partial_sum/np.log(2)
print("Deviation:", percentace_deviation, "%")
```
